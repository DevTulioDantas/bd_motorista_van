Tabela Motorista

```python
CREATE TABLE MOTORISTA (
    COD_MOTORISTA NUMBER PRIMARY KEY,
    PLACA_PERUA NUMBER(4),
    MODELO NUMBER(4),
    CAPACIDADE NUMBER,
    CPF NUMBER(4),
    TELEFONE NUMBER(4),
    EMAIL VARCHAR2(100), -- Suponho que EMAIL deve ser VARCHAR2, no NUMBER
    CONTA_PAYPAL VARCHAR2(50) -- Suponho que CONTA_PAYPAL deve ser VARCHAR2, no NUMBER
);
```

Tabela Escola

```python
CREATE TABLE ESCOLA (
    COD_ESCOLA NUMBER PRIMARY KEY,
    NOME VARCHAR2(100), -- Suponho que NOME deve ser VARCHAR2, no NUMBER
    ENDERECO VARCHAR2(200) -- Suponho que ENDERECO deve ser VARCHAR2, no NUMBER
);
```

Tabela SERVICO

```python
CREATE TABLE SERVICO (
    COD_SERVICO NUMBER PRIMARY KEY,
    COD_ALUNO NUMBER,
    COD_CLIENTE NUMBER,
    COD_MOTORISTA NUMBER,
    CONSTRAINT fk_servico_aluno FOREIGN KEY (COD_ALUNO) REFERENCES ALUNO(COD_ALUNO),
    CONSTRAINT fk_servico_cliente FOREIGN KEY (COD_CLIENTE) REFERENCES CLIENTE(COD_CLIENTE),
    CONSTRAINT fk_servico_motorista FOREIGN KEY (COD_MOTORISTA) REFERENCES MOTORISTA(COD_MOTORISTA)
);
```

Tabela CLIENTE

```python
CREATE TABLE CLIENTE (
    COD_CLIENTE NUMBER PRIMARY KEY,
    COD_ALUNO NUMBER,
    NOME VARCHAR2(100), -- Suponho que NOME deve ser VARCHAR2, no NUMBER
    ENDERECO VARCHAR2(200), -- Suponho que ENDERECO deve ser VARCHAR2, no NUMBER
    CPF NUMBER(4),
    EMAIL VARCHAR2(100), -- Suponho que EMAIL deve ser VARCHAR2, no NUMBER
    CONTA_PAYPAL VARCHAR2(50), -- Suponho que CONTA_PAYPAL deve ser VARCHAR2, no NUMBER
    CONSTRAINT fk_cliente_aluno FOREIGN KEY (COD_ALUNO) REFERENCES ALUNO(COD_ALUNO)
);
```

Tabela ALUNO

```python
CREATE TABLE ALUNO (
    COD_ALUNO NUMBER PRIMARY KEY,
    COD_MOTORISTA NUMBER,
    COD_CLIENTE NUMBER,
    COD_ESCOLA NUMBER,
    NOME VARCHAR2(100), -- Suponho que NOME deve ser VARCHAR2, no NUMBER
    ENDERECO VARCHAR2(200), -- Suponho que ENDERECO deve ser VARCHAR2, no NUMBER
    CONSTRAINT fk_aluno_motorista FOREIGN KEY (COD_MOTORISTA) REFERENCES MOTORISTA(COD_MOTORISTA),
    CONSTRAINT fk_aluno_cliente FOREIGN KEY (COD_CLIENTE) REFERENCES CLIENTE(COD_CLIENTE),
    CONSTRAINT fk_aluno_escola FOREIGN KEY (COD_ESCOLA) REFERENCES ESCOLA(COD_ESCOLA)
);
```
